# Two types of cell
- Python code cell. code can be evaluated by typing shift-return
- Markup cell. Easy outling, Latex source

# Two editing modes in the notebook
- notebook editing mode(Blue vertical bar at left of cell)
    - 'y' key - change current cell to Python
    - 'm' key - change current cell to Markup
    - 'a' key - add a cell above current one
    - 'b' key - add a cell below current one
    - 'x' key - kill current cell
    - 'command-s' keys - save notebook
    - 'return' key or double click inside cell - enter cell editing mode
    
- cell editing mode(Green vertical bar at left of cell)
    - edit Python code or Markup source
    - knows basic Emacs key bindings, c-a,c-e, etc
    - 'shift-return' keys - evaluate cell(run python code, or render markup)
    - esc key - enter notebook editing mode
    
- Menubar has useful functionality, including
    - File
        - new notebook, open, rename
    - Edit
        - Split cell, Move Cell Up, Move Cell down
    - Cell
        - All output|Clear
    - Kernel
        - Interrupt(try to stop computation)
        - Restart(get a fresh Python - you will lose all your variable values)

# Lorenz attractor
- Solve differential equation numerically and plot

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

def make_lorenz(sigma, r, b):
    def func(statevec, t):
        x, y, z = statevec
        return [ sigma * (y - x), r * x - y - x * z, x * y - b*z]
    return func

lorenz_eq = make_lorenz(10., 28., 8./3.)
tmax = 50
tdelta = 0.005
tvalues = np.arange(0, tmax, tdelta)
ic = np.array([0.0, 1.0, 0.0])
sol = odeint(lorenz_eq, ic, tvalues)
x, y, z = np.array(list(zip(*sol)))
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=1, color='red')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
plt.show()
```

```python
from numpy import *
import pylab

def mandelbrot( h,w, maxit=20 ):
        '''Returns an image of the Mandelbrot fractal of size (h,w).
        '''
        y,x = ogrid[ -1.4:1.4:h*1j, -2:0.8:w*1j ]
        c = x+y*1j
        z = c
        divtime = maxit + zeros(z.shape, dtype=int)

        for i in range(maxit):
                z  = z**2 + c
                diverge = z*conj(z) > 2**2            # who is diverging
                div_now = diverge & (divtime==maxit)  # who is diverging now
                divtime[div_now] = i                  # note when
                z[diverge] = 2                        # avoid diverging too much

        return divtime

pylab.imshow(mandelbrot(400,400,40))
pylab.show()
```
